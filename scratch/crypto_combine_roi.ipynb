```python
import csv
import pandas as pd
```

```python
lin_roi_df = pd.read_csv("crypto_lin_roi.csv",parse_dates=True, index_col="timestamp")
lin_roi_df.index = pd.to_datetime(lin_roi_df.index)

grad_roi_df = pd.read_csv("crypto_grad_roi.csv",parse_dates=True, index_col="timestamp")
grad_roi_df.index = pd.to_datetime(grad_roi_df.index)

imb_roi_df = pd.read_csv("crypto_imb_roi.csv",parse_dates=True, index_col="timestamp")
imb_roi_df.index = pd.to_datetime(imb_roi_df.index)

xgb_roi_df = pd.read_csv("crypto_xgb_roi.csv",parse_dates=True, index_col="timestamp")
xgb_roi_df.index = pd.to_datetime(xgb_roi_df.index)
```

```python
combined_df = pd.concat([lin_roi_df, grad_roi_df,imb_roi_df,xgb_roi_df], axis=1)
```

```python
combined_df.to_csv("crypto_combined_roi.csv")
```
